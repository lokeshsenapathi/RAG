```python
data_text=loader_dir[-1].page_content
data_text
```

```python
text_spliter = CharacterTextSplitter(
    separator='\n',
    chunk_size= 132,
    chunk_overlap=10,
    length_function = len
)

tesxt_spillter = text_spliter.split_text(data_text)
print(tesxt_spillter[0])
print("------------------------------------")
print(tesxt_spillter[1])
print("------------------------------------")
print(tesxt_spillter[2])
```

```python
rec_spliter = RecursiveCharacterTextSplitter(
    separators=['\n','\n\n',' ','',',','.'],
    chunk_size= 80,
    chunk_overlap=20,
    length_function = len
)

rec_test_splitter = rec_spliter.split_text(data_text)

print(rec_test_splitter[0])
print(rec_test_splitter[1])
print(rec_test_splitter[2])
print(rec_test_splitter[3])
print(rec_test_splitter[4
                        ])
```
